```python
import pandas as pd
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
dir(digits)
```

```python
df=pd.DataFrame(digits.data)
```

```python
df['target']=digits.target
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(df.drop('target',axis='columns'),df['target'],test_size=0.23)
```

```python
clf=SVC(C=3)
```

```python
clf.fit(xtrain,ytrain)
```

```python
clf.score(xtest,ytest)
```
